```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('bjjheroes_matches.csv', index_col = 0)
def make_binary(string):
    if string == 'W':
        return 1
    else:
        return 0
df['outcome_binary'] = df['outcome'].apply(make_binary)

df.head(20)
```

```python
total_matches = df.groupby('name').agg(len)['method']
win_rate = df.groupby('name')['outcome_binary'].mean()
total_matches.rename('total_matches')
win_rate.rename('win_rate')
```

```python
new_df = pd.concat([total_matches, win_rate], axis = 1)
new_df.columns = ['match_count', 'win_rate']
new_df
```

```python
sns.pairplot(new_df)
new_df.corr()
```

```python
all_outcomes = pd.Series(df.method.unique())
df.method = df.method.map(lambda x: 'NA' if type(x) == float else x)
```

```python
def is_submission(method):
    decision_methods = ['Pts', 'Decision', 'Points', 'NA', 'N/A', 'DQ', 'Adv']
    if any([(decision in method) for decision in decision_methods]):
        return 0
    else:
        return 1
method = 'DQ'

is_submission(method)
```

```python
df['by_sub'] = df.method.map(is_submission)
```

```python
df[df.name == 'AJ Agazarm']['by_sub'].mean()
```

```python
sub_rate = df[df.outcome_binary == 1].groupby('name').by_sub.mean()
```

```python
new_df['sub_rate'] = sub_rate
```

```python
sns.pairplot(new_df)
```

```python
new_df.corr()
```

```python
data = new_df.sub_rate.dropna()

fig, ax = plt.subplots(1,1)

ax.boxplot(data, vert = False)
```

```python
sub_loss_rate = df[df.outcome_binary == 0].groupby('name').by_sub.mean()
new_df['sub_loss_rate'] = sub_loss_rate
```

```python
sns.pairplot(new_df)
```
